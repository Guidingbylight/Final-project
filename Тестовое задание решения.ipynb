```python
#3 вопрос

df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные об аудитории')

df['date'] = pd.to_datetime(df['date'])

first_day = df[df['date'] == '2023-11-01']['user_id'].unique()

second_day = df[df['date'] == '2023-11-02']['user_id'].unique()

returned = set(first_day) & set(second_day)

retention = len(returned) / len(first_day)

print('Retention D1:', retention * 100)
```

```python
#8 вопрос

import pandas as pd
from scipy import stats

df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные АБ тестов')

print(df.head())
print(df['experiment_num'].unique())

results = []

for exp in df['experiment_num'].unique():
    temp = df[df['experiment_num'] == exp]
    
    test = temp[temp['experiment_group'] == 'test']['revenue']
    control = temp[temp['experiment_group'] == 'control']['revenue']
    
    t_stat, p_value = stats.ttest_ind(test, control, equal_var=False)
    
    arpu_test = test.mean()
    arpu_control = control.mean()
    
    results.append({
        'experiment': exp,
        'ARPU_test': arpu_test,
        'ARPU_control': arpu_control,
        'p_value': p_value
    })

results_df = pd.DataFrame(results)
print(results_df)
```
